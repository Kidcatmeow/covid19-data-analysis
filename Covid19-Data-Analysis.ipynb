### 6. The rate of COVID-19 infection among patients compared to the general population

```python
pop_df = pd.read_csv('thai_population_63.csv')
pop_df
```

```python
pop_df = pop_df [['จังหวัด','รวม']]
pop_df.head()
```

```python
df.groupby('province_of_isolation').count()['No.']
```

```python
#Reset Index
province_df = df.groupby('province_of_isolation').count()['No.'].reset_index()
province_df
```

```python
#Replace duplicate provinces (due to abbreviations)
df['province_of_isolation'].replace('กทม','กรุงเทพมหานคร') 
```

```python
df.insert(2,'province',df['province_of_isolation'].replace('กทม','กรุงเทพมหานคร') ,True)
```

```python
df.groupby('province').count()['No.']
```

```python
#Reset Index
province_df = df.groupby('province').count()['No.'].reset_index()
province_df
```

```python
#merge population df with province df 
prov_pop_df = pd.merge(province_df,pop_df,left_on='province',right_on='จังหวัด')
prov_pop_df
```

```python

prov_pop_df['rate'] = prov_pop_df['No.'] * 100000 / prov_pop_df['รวม']
prov_pop_df
```

```python
# 10 provinces with the lowest percentage of patients / normal population
top10_prov_df = prov_pop_df.sort_values('rate',ascending=False).head(10).set_index('จังหวัด')
top10_prov_df 
```

```python
top10_prov_df[['No.','rate']].plot.bar(figsize=(10,5))
```
